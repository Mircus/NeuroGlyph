# NeuroGlyph Parser Demo
Questo notebook dimostra come usare il parser NeuroGlyph v3.7+ per:
- Parsare blocchi di testo in linguaggio NeuroGlyph
- Ottenere una struttura interna (dizionario)
- Renderizzare di nuovo il testo da struttura interna

```python
# Importa il parser
from neuroglyph_parser import NeuroGlyphParser
```

```python
# Istanzia il parser
parser = NeuroGlyphParser()
```

```python
# Esempio base di blocco NeuroGlyph
example_block = """
/mind: Eidos; Logos
/act: define
/focus: Truth
/context: relational_epistemology
/deliverable: diagram
"""
```

```python
# Parsing del blocco
parsed = parser.parse_block(example_block)
parsed
```

```python
# Rendering inverso (ricostruzione NG)
rendered = parser.render(parsed)
print(rendered)
```
